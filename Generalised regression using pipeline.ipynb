```python
# importing libraries
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np

from sklearn.preprocessing import PolynomialFeatures
from sklearn.linear_model import LinearRegression
from sklearn.pipeline import Pipeline
from sklearn import metrics

from sklearn.model_selection import train_test_split
```

```python
# loading the csv file into a dataframe
df= pd.read_csv('total-electricity-consumption-us.csv')
df.head()
```

```python
# check the number of missing values
df.info()
```

```python
sns.lineplot(x='Year',y='Consumption',data=df)
plt.grid()
```

```python
X = df['Year']
y = df['Consumption']

x_train,x_test,y_train,y_test=train_test_split(X,y,test_size = 0.4)
```

```python
x_train.values.reshape(-1,1).shape
```

```python
y_train.values.reshape(-1,1).shape
```

```python
# pipeline for degree 1
pipeline1 = Pipeline([('poly_feature', PolynomialFeatures(degree=1)),('model',LinearRegression())])
pipeline1.fit(x_train.values.reshape(-1,1),y_train.values.reshape(-1,1))

x_test_pred1=pipeline1.predict(x_test.values.reshape(-1,1))

sns.scatterplot(x=x_test,y=y_test,color='r')
sns.lineplot(x=x_test,y=x_test_pred1.reshape(-1),color='g' )
```

```python
# pipeline for degree 2
pipeline2 = Pipeline([('poly_feature', PolynomialFeatures(degree=2)),('model',LinearRegression())])
pipeline2.fit(x_train.values.reshape(-1,1),y_train.values.reshape(-1,1))

x_test_pred2=pipeline2.predict(x_test.values.reshape(-1,1))

sns.scatterplot(x=x_test,y=y_test,color='b')
sns.lineplot(x=x_test,y=x_test_pred2.reshape(-1),color='black' )
```

```python
# pipeline for degree 3
pipeline3 = Pipeline([('poly_feature', PolynomialFeatures(degree=3)),('model',LinearRegression())])
pipeline3.fit(x_train.values.reshape(-1,1),y_train.values.reshape(-1,1))

x_test_pred3=pipeline3.predict(x_test.values.reshape(-1,1))

sns.scatterplot(x=x_test,y=y_test,color='grey')
sns.lineplot(x=x_test,y=x_test_pred3.reshape(-1),color='orange' )
```

```python
# pipeline for degree 4
pipeline4 = Pipeline([('poly_feature', PolynomialFeatures(degree=4)),('model',LinearRegression())])
pipeline4.fit(x_train.values.reshape(-1,1),y_train.values.reshape(-1,1))

x_test_pred4=pipeline4.predict(x_test.values.reshape(-1,1))

sns.scatterplot(x=x_test,y=y_test,color='grey')
sns.lineplot(x=x_test,y=x_test_pred4.reshape(-1),color='orange' )
```

```python
plt.subplot(2,2,1)
sns.scatterplot(x=x_test,y=y_test,color='r')
sns.lineplot(x=x_test,y=x_test_pred1.reshape(-1),color='g' )

plt.subplot(2,2,2)
sns.scatterplot(x=x_test,y=y_test,color='b')
sns.lineplot(x=x_test,y=x_test_pred2.reshape(-1),color='black' )

plt.subplot(2,2,3)
sns.scatterplot(x=x_test,y=y_test,color='grey')
sns.lineplot(x=x_test,y=x_test_pred3.reshape(-1),color='orange' )

plt.subplot(2,2,4)
sns.scatterplot(x=x_test,y=y_test,color='grey')
sns.lineplot(x=x_test,y=x_test_pred4.reshape(-1),color='orange' )

plt.tight_layout()
```
